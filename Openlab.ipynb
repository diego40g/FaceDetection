```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar el archivo CSV
data = pd.read_excel('base.xlsx')

# Seleccionar las variables independientes (X) y la variable dependiente (y)
selected_features = ['p5_10', 'p5_11', 'p7_1', 'p7_2', 'p7_3', 'p7_4', 'p7_5', 'p7_6', 'p7_7']
X = data[selected_features]
y = data['p8']

# Aplicar transformación logarítmica a la variable objetivo
y = np.log(y)

# Crear el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Obtener los coeficientes y el intercepto
coeficientes = model.coef_
intercepto = model.intercept_

# Mostrar los resultados
print("Coeficientes:", coeficientes)
print("Intercepto:", intercepto)

# Crear una figura y un subplot
fig, ax = plt.subplots()

# Graficar los datos reales
ax.scatter(X['p5_10'], y, color='blue', label='Datos reales')

# Graficar la línea de regresión
x_line = list(range(X['p5_10'].min(), X['p5_10'].max()+1))
y_line = model.predict(pd.DataFrame({'p5_10': x_line}))
ax.plot(x_line, y_line, color='red', label='Regresión lineal')

# Configurar el título y las etiquetas de los ejes
ax.set_title('Regresión Lineal')
ax.set_xlabel('p5_10')
ax.set_ylabel('log(p8)')

# Mostrar la leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
